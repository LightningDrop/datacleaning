# Removing Rows with No Data Input
The goal is to delete rows that have multiple empty fields in the dataset. A large number of rows have numerous questions/columns that are unfilled; and the task is to delete these rows that present multiple missing data to leave out the filled rows.


```python
#import pandas
import pandas as pd
```

```python
#read dataset file
data = pd.read_csv('./candyhierarchy2017.csv', encoding = "ISO-8859-1")
```

Start of the process: The task is to removes rows in the dataset that have multiple missing values. Rows that have several missing fields will not be included in the finalized data.

```python
#show the first 5 rows of the data to check contents
data.head()
```

```python
#drop rows base on their index value.
#To test this, drop index 0 first since that row do not have any record.
data = data.drop(0)
```

```python
#Show and check if the row is deleted
data.head()
```

Begin dropping rows: upon observation, rows that have multiple missing data starts to miss input in the column  "Q6 | Any full-sized candy bar," so this column is used as a "filter" to drop these rows.

```python
data = data.dropna(subset = ['Q6 | Any full-sized candy bar'])
```

```python
data.head()
```

```python
#create a new updated candyfile dataset
data.to_csv('updated_candy_set_rows.csv')
```
